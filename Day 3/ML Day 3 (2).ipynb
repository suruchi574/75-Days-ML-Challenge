### Dummy Variables

Dummy variables are qualitative variables or discrete variables that represent categorical data and can take the values as 0 or 1 to indicate the absence or presence of a specified attribute respectively.

Dummy variables are also known as indicator variables, design variables, and binary basis variables.

###### Creating dummy variable for town values that have data in text 

###### Code

```python
import pandas as pd 
df = pd.read_csv("homeprices1.csv")
df
```

```python
dummies= pd.get_dummies(df.town)
# we are creating dummy variable for town values because our linear regression modl can't operate in text values . so in place of text we 'll set true /false '
dummies
```

```python
merged = pd.concat([df,dummies],axis ='columns')
merged
```

```python
final = merged.drop(['town','west windsor'],axis='columns')
#As i have my town in numerical value(true/false) we don't need to keep it in our data frame so we'll drop it
# And we'll also drop one dummy variable so in future they can't create conflict 
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price',axis='columns')
```

```python
x
```

```python
y= final.price
```

```python
y
```

```python
model.fit(x.values,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x.values,y)
```
